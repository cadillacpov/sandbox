# Task 2

```python
# Drop rows where any column has NaN values
df_drop = df.dropna()
print("Data Frame after dropping rows with missing values:")
print(df_drop)
```

```python
# Fill missing values with default values
df_fill = df.copy()
# Fill numerical columns with 0
df_fill['Age'] = df_fill['Age'].fillna(0)
df_fill['Salary'] = df_fill['Salary'].fillna(0)
# Fill categorical columns with 'Unknown'
df_fill['Country'] = df_fill['Country'].fillna('Unknown')
df_fill['Purchased'] = df_fill['Purchased'].fillna('Unknown')
print("Data Frame after filling missing values:")
print(df_fill)
```

```python
# Check for missing values in df_filled (True = missing, False = present)
print("Missing Values in df_filled (True = Missing):")
print(df_fill.isnull())

print("\nCount of missing values per column after filling:")
print(df_fill.isnull().sum())
```

```python
# Fill missing Age values with the mean of the Age column (in-place)
df['Age'] = df['Age'].fillna(df['Age'].mean())
print("Data Frame after filling Age with mean:")
print(df)
```

```python
# Fill missing Salary values with the median of the Salary column
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
print("Data Frame after filling Salary with median:")
print(df)
```

```python
# Fill missing Purchased values with the mode (most frequent value) of the column
df['Purchased'] = df['Purchased'].fillna(df['Purchased'].mode()[0])
print("Data Frame after filling Purchased with mode:")
print(df)
```

```python
# Apply forward fill to the Country column only
df['Country'] = df['Country'].ffill()
print("Data Frame after applying forward fill to Country:")
print(df)
```
